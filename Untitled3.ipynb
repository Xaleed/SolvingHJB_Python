```python
@numba.jit(nopython = True) 
def DGR1(b1, j):
    return -exp(-lambda1*j*DELTA/b1)+exp(-lambda1*(j-1)*DELTA/b1)
```

```python
def IR(i,b1,b2,b3):
    j = np.arange(i+1)
    return np.sum(d[i-j]*DGR(b1, b2, b3, j))
```

```python
@numba.jit(nopython = True) 
def E(b1): 
    return b1/lambda1
```

```python
E1 = E(1)
@numba.jit(nopython = True) 
def pR1(b1):
    return (1+eta)*beta1*E1 - ((1+eta1)*beta1*(E1-E(b1)))
```

```python
for i in range(n):
    def H(b):
        pRR = pR1(b[0])
        j1 = np.arange(i)
        j2 = np.arange(1,i+1)
        if pRR>0: return ((c+beta1)*d[i]-0.5*beta1*sum((d[i-j1]+d[i-j2])*DGR1(b[0], j2)))/pRR
        else : return  np.inf
    MIN = differential_evolution(H, [(0.000001, 1.0)])#,popsize=30, strategy='rand1exp')
    dhat[i] = MIN.fun
    d[i+1]=d[i]+DELTA*dhat[i]

    B1[i] = MIN.x[0]
    #B2[i] = MIN.x[1]
   # B3[i] = MIN.x[2]
    if i%(n//20)==0: 
        print(str(round(100*i/n))+'%', end=', ')
        print('d('+str(round(i*DELTA,2))+') = '+str(d[i]))
print('100%')
```
